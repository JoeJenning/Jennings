```python
import matplotlib.pyplot as plt
import numpy as np
import os
import scipy.io

X = np.random.rand(100,100)
U, S, VT = np.linalg.svd(X,full_matrices=0)

n,m=X.shape

plt.figure(1)
plt.semilogy(S)
plt.title('Singular Values')
plt.show()
```

```python
S_big=np.zeros((100, 100))
means=np.zeros(n)
medians=np.zeros(n)

fig = plt.figure(figsize=(16, 12))
gs = plt.GridSpec(nrows=2, ncols=2)

for i in range(100):
    X = np.random.rand(100,100)
    U, S_big[:,i], VT = np.linalg.svd(X,full_matrices=0)
    
ax0=fig.add_subplot(gs[0,:])
ax0.boxplot(S_big)
ax0.set_title('Singular Values')

for r in range(n):
    means[r]=np.mean(S_big[:r,0])
    medians[r]=np.median(S_big[:r,0])

ax1=fig.add_subplot(gs[1,0])
ax1.plot(means)
ax1.set_title('Mean')

ax2=fig.add_subplot(gs[1,1])
ax2.plot(medians)
ax2.set_title('Median')
plt.show()
```

```python
for j in (50,200):
    S_big=np.zeros((j, j))
    means=np.zeros(j)
    medians=np.zeros(j)

    fig = plt.figure(figsize=(16, 12))
    gs = plt.GridSpec(nrows=2, ncols=2)

    for i in range(j):
        X = np.random.rand(j,j)
        U, S_big[:,i], VT = np.linalg.svd(X,full_matrices=0)

    ax0=fig.add_subplot(gs[0,:])
    ax0.boxplot(S_big)
    ax0.set_title('Singular Values '+ str(j) + 'x'+str(j))

    for r in range(j):
        means[r]=np.mean(S_big[:r,0])
        medians[r]=np.median(S_big[:r,0])

    ax1=fig.add_subplot(gs[1,0])
    ax1.plot(means)
    ax1.set_title('Mean '+ str(j) + 'x'+str(j))

    ax2=fig.add_subplot(gs[1,1])
    ax2.plot(medians)
    ax2.set_title('Median '+ str(j) + 'x'+str(j))
    plt.show()
```
