```python
import matplotlib.pyplot as plt
import numpy as np
import os
plt.rcParams['figure.figsize'] = [16, 8]
plt.rcParams.update({'font.size':18})

L=2
R=1

fig,ax = plt.subplots()

#define dart board
rectangle = plt.Rectangle((-1,-1),L, L)
ax.add_patch(rectangle)
circle = plt.Circle((0,0),R, fc='red')
ax.add_patch(circle)
ax.axis('scaled')
```

```python
x=(np.random.rand(2,10000)*2)-1

radii=np.sqrt(np.square(x[0])+np.square(x[1]))
count=0
for i in range(radii.shape[0]):
    if radii[i]<1:
        count+=1
        
print("Expected number of darts with r<1: "+str((np.square(L)-(np.pi*np.square(R)))*10000)+"/10000")
print("Actual: "+str(count)+"/10000")
```

```python
radii=np.zeros((9,10000))
count=np.zeros(9)
for N in range(2,11):
    x=(np.random.rand(N,10000)*2)-1
    s=np.zeros(10000)
    for i in range(N):
        s=s+np.square(x[i])
    radii[N-2]=np.sqrt(s)
    
    for j in range(10000):
        if radii[N-2,j]<1:
            count[N-2]+=1
plt.plot(count)
plt.show()
```
